```python
from civetqc.data.dataset import Dataset
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import fbeta_score, classification_report, balanced_accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
```

```python
class Model:
    
    def __init__(self, data, clf) -> None:
        self.data = data
        self.clf = clf.fit(data.train["features"], data.train["target"])
        self.predicted = self.clf.predict(data.test["features"])
    
    def __str__(self):
        return (
            f"f2_score: {self.f2_score}\n"
            f"balanced_accuracy_score: {self.balanced_accuracy_score}"
        )
    
    @property
    def f2_score(self):
        return fbeta_score(self.data.test["target"], self.predicted, beta=2)
    
    @property
    def classification_report(self):
        return classification_report(self.data.test["target"], self.predicted)
    
    @property
    def balanced_accuracy_score(self):
        return balanced_accuracy_score(self.data.test["target"], self.predicted) 
```
