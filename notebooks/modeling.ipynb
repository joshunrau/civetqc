```python
import os

from civetqc.data.datasets import Dataset
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
studies_dir = "/Users/joshua/Developer/civetqc/data"

study_paths = [
    (
        os.path.join(studies_dir, "FEP", "FEP_civet_data.csv"), 
        os.path.join(studies_dir, "FEP", "FEP_QC.csv")
    ),
    (
        os.path.join(studies_dir, "LAM", "LAM_civet_data.csv"), 
        os.path.join(studies_dir, "LAM", "LAM_QC.csv")
    ),
    (
        os.path.join(studies_dir, "INSIGHT", "INSIGHT_civet_data.csv"), 
        os.path.join(studies_dir, "INSIGHT", "INSIGHT_QC.csv")
    ),
    (
        os.path.join(studies_dir, "TOPSY", "TOPSY_civet_data.csv"), 
        os.path.join(studies_dir, "TOPSY", "TOPSY_QC.csv")
    )
]
```

```python
class Model(Dataset):
    
    def __init__(self, clf, study_paths: list) -> None:
        super().__init__(study_paths)
        self.clf = clf.fit(self.train.features, self.train.target)
        self.predicted = self.clf.predict(self.test.features)
    
    @property
    def classification_report(self):
        return classification_report(self.test.target, self.predicted)
    
    @property
    def accuracy_score(self):
        return accuracy_score(self.test.target, self.predicted)
```

```python
model = Model(RandomForestClassifier(n_estimators=100, random_state=0), study_paths)
```
