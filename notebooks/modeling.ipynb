```python
class Model(Dataset):
    
    def __init__(self, clf, study_paths: list, balanced: bool = False) -> None:
        super().__init__(study_paths, balanced)
        self.clf = clf.fit(self.train.features, self.train.target)
        self.predicted = self.clf.predict(self.test.features)
    
    @property
    def f2_score(self):
        return fbeta_score(self.test.target, self.predicted, average='binary', beta=2)
```

```python
model = Model(RandomForestClassifier(n_estimators=100, random_state=0), study_paths, True)
model.f2_score
```
